```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(tf.__version__)
```

```python
obs = 1000
xs = np.random.uniform(low=-10,high=10,size = (obs,1))
zs = np.random.uniform(low=-10,high=10,size = (obs,1))

generaed_inputs = np.column_stack((xs,zs))

noise = np.random.uniform(-1,1,(obs,1))

generated_targets = 2*xs-3*zs+5+noise

np.savez('TF_intro',inputs = generaed_inputs, targets = generated_targets )
```

```python
trading_data = np.load('TF_intro.npz')
```

```python
input_size = 2
output_size = 1
```

```python
model = tf.keras.Sequential([
                            tf.keras.layers.Dense(output_size)
                            ])

model.compile(optimizer = 'sgd',loss = 'mean_squared_error' )
model.fit(trading_data['inputs'],trading_data['targets'], epochs=1000,verbose=2)
```

```python
plt.plot(np.squeeze(model.predict_on_batch(trading_data['inputs'])),np.squeeze(trading_data['targets']))
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```

```python
model = tf.keras.Sequential([
                            tf.keras.layers.Dense(output_size,
                                                    kernel_initializer = tf.random_uniform_initializer(minval = -0.1, maxval = 0.1),
                                                    bias_initializer = tf.random_uniform_initializer(minval = -0.1, maxval = 0.1)
                                                 )
                            ])
custom_optimizer = tf.keras.optimizers.SGD(learning_rate=0.02)

model.compile(optimizer = custom_optimizer,loss = 'mean_squared_error' )
model.fit(trading_data['inputs'],trading_data['targets'], epochs=1000,verbose=2)
```
